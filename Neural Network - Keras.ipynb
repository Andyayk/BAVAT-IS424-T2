```python
# Importing the libraries
import keras
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read from dataframe
df = pd.read_pickle("explored_data")
df = df.replace([np.inf, -np.inf, np.nan], 0) #removing infinite/nan values

# Check the columns using dtypes
print(df.dtypes)
# Randomly sample 5 records with .sample(5)
df.sample(5)
```

```python
#movies = pd.read_csv("confusionmatrixtest2.csv")
#X = movies.drop('revenue', axis=1)
X = df[['budget', 'weekday', 'day', 'month', 'year', 'runtime', 'weighted_rating']]
y = df[['bin']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
classifier = Sequential()
```

```python
classifier.add(Dense(units = 7, kernel_initializer = 'uniform', activation = 'relu', input_dim = 7))
classifier.add(Dense(units = 7, kernel_initializer = 'uniform', activation = 'relu', input_dim = 7))
classifier.add(Dense(units = 7, kernel_initializer = 'uniform', activation = 'relu', input_dim = 7))
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'relu', input_dim = 7))

classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['mae', 'accuracy'])
```

```python
print(y.shape)
print(X.shape)
```

```python
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

```python
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

def printModelAccuracy(y_test, y_pred):
    # Find the confusion matrix of the result
    cm = pd.DataFrame(confusion_matrix(y_test, y_pred, labels=[1, 2, 3, 4, 5]), \
        index=['true:1', 'true:2', 'true:3', 'true:4', 'true:5'], 
        columns=['pred:1', 'pred:2', 'pred:3', 'pred:4', 'pred:5'])
    print("Confusion Matrix:")
    print(cm)

    # Find the accuracy and F1 score of the result
    asr = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='macro')
    print("Accuracy:", asr)
    print("F1:", f1)
    
printModelAccuracy(y_test, y_pred)
```
